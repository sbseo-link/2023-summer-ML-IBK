```python
data_dir = "../src/spam"

# read all files with '.txt' extension in data_dir
import glob
filenames = glob.glob(os.path.join(data_dir, "*.txt"))
print(filenames)
```

```python
# read all files into a list
texts = []
for filename in filenames:
    # open file
    with open(filename, "r") as f:
        # read file content
        text = f.read()
        # append to list
        texts.append(text)

# print texts
print(texts)
```

```python
len(texts)
```

```python
# remove 'new line character' and 'tab character'
texts_without_chars = [text.replace("\n", " ") for text in texts]
print(texts)
```

```python
# split texts into words
texts_into_words = [text.split(" ") for text in texts]
print(texts_into_words)
```

```python
len(texts)
```

```python
# collect all words into a list
all_words = []
for text in texts_into_words:
    for word in text:
        all_words.append(word)
print(all_words)
```

```python
from konlpy.tag import Kkma

def extract_nouns(text):
    kkma = Kkma()
    nouns = kkma.nouns(text)
    return nouns
```

```python
extract_nouns("안녕하세요. 저는 홍길동입니다.")
```

```python
# extract nouns from 'all_words' and make new list
all_nouns = []
for text in all_words:
    nouns = extract_nouns(text)
    all_nouns.extend(nouns)

print(all_nouns)
```

```python
# iterate over all_nouns and count each noun using Counter and make a dictionary with nouns as keys and counts as values
from collections import Counter
noun_counter = Counter(all_nouns)
print(noun_counter)
```

```python
# sort noun_counter by values in descending order
noun_counter_sorted = noun_counter.most_common()
```

```python
noun_counter_sorted
```

```python
# extract words with count more than 10
noun_list = []
for noun, count in noun_counter_sorted:
    if count > 10:
        noun_list.append(noun)

print(noun_list)
```

```python
# remove words in noun_list if it is a string type of integer
noun_list_without_int = [noun for noun in noun_list if not noun.isdigit()]
print(noun_list_without_int)
```
