```python
# 필요한 라이브러리 import
import requests
```

```python
# url 정의 및 요청
url = 'https://api-testnet.bybit.com/spot/v3/public/quote/ticker/price'
response = requests.get(url)
```

```python
# 결과 상태 확인
response.status_code
```

```python
# 결과 내용 확인
response.content
```

```python
# 결과 내용 확인
spot_data = response.json()
```

```python
type(spot_data)
```

```python
spot_data
```

```python
# 필요 정보 추출
spot_data = spot_data['result']['list']
spot_data
```

```python
# 필요 정보 추출
spot_BTCUSDT = None
for ticker in spot_data:
  if ticker['symbol'] == 'BTCUSDT':
    spot_BTCUSDT = ticker
```

```python
spot_BTCUSDT
```

```python
# url 정의 및 요청
url = 'https://api-testnet.bybit.com/derivatives/v3/public/tickers'
response = requests.get(url)
```

```python
deriv_data = response.json()['result']['list']
```

```python
deriv_data
```

```python
# 필요 정보 추출
deriv_BTCUSDT = None
for ticker in deriv_data:
  if ticker['symbol'] == 'BTCUSDT':
    deriv_BTCUSDT = ticker
```

```python
deriv_BTCUSDT
```

```python
# 필요 정보 추출
final_data = {'spot_last_price': spot_BTCUSDT['price'], 'deriv_last_price': deriv_BTCUSDT['lastPrice'] }
final_data
```

```python
# 필요 정보 계산
final_data['spread'] = (float(final_data['deriv_last_price']) - float(final_data['spot_last_price'])) / float(final_data['spot_last_price']) * 100
final_data
```
