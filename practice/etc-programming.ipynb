```python
start_idx, end_idx = 0, 0
infos = {}
for _, row in information.iterrows():
    field_name = row['항목명']
    length = row['길이']
    accum_length = row['누적\n길이']
    end_idx = start_idx + length
    print(field_name, start_idx, end_idx)
    infos[field_name] = [start_idx, end_idx]
    start_idx = end_idx
```

```python
# 233_38_231_12-10022.log 파일을 순회하면서 종목코드 'KR7005360003' 데이터 출력
with open('../src/233_38_231_12-10022.log', 'rb') as f:
    total_data = f.readlines()
    
for line in total_data:
    if line[17:29] == b'KR7005360003':
        print(line)
```

```python
# 각각의 거래대금 출력
for line in total_data:
    if line[17:29] == b'KR7005360003':
        price = float(line[59:70].decode('euc-kr'))
        qty = float(line[70:80].decode('euc-kr'))
        total_krw = price * qty
        print('거래대금', total_krw)
```

```python
# 각각의 거래대금 100만원 이상 출력
for line in total_data:
    if line[17:29] == b'KR7005360003':
        price = float(line[59:70].decode('euc-kr'))
        qty = float(line[70:80].decode('euc-kr'))
        total_krw = price * qty
        if total_krw >= 1000000:
            print('거래대금', total_krw)
        
```

```python
# 각각의 거래대금 100만원 이상 체결가가 직전가격보다 높으면 매수
pivot_price = 0
for line in total_data:
    if line[17:29] == b'KR7005360003':
        cur_price = float(line[59:70].decode('euc-kr'))
        cur_qty = float(line[70:80].decode('euc-kr'))
        total_krw = cur_price * cur_qty
        if total_krw >= 1000000:
            if (cur_price > pivot_price) & (pivot_price != 0):
                print('거래대금:', total_krw, '직전가격:', pivot_price, '현재가격:', cur_price, '매수')
                
        pivot_price = cur_price
        
```
